```python
# importando as blibiotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv")
df.drop('Unnamed: 0',axis=1, inplace=True) # excluir uma coluna que não será necessária
```

```python
df.head() # carregar as cinco primeiras linha do dataframe
```

Neste dataset temos as informações sobre o diamante e o seu determinado preço

```python
df.describe()
```

## Amostragem

```python
amostra1 = df.iloc[0:3000,:]
amostra1
```

```python
amostra1.describe()
```

Comparando primeiramente a primeira amostra em relação ao df vemos que não é tão representativa pois além da media não srem proximas, o desvio padrão da amostra 1 mostra que os valores estão mais próximos se comparado com a população (vendo a variavel do preço por exemplo).

```python
amostra2 = df.iloc[3001:6000,:]
amostra2
```

```python
amostra2.describe()
```

Da mesma forma que vimos na primeira amostra temos agora com a amostra 2

```python
amostra3 = df.sample(3000, random_state=1)
amostra3
```

```python
amostra3.describe()
```

Observa-se se comparamos a amostra 3 tem valores próximos ao da população o que pode indicar que essa pode ser uma amostra representativa, além da variavel preço, temos outras variaveis que corroboram com isso

### Observando a amostra por meio de gráficos

Agora vamos observar por meio de gráfico, vamos observar pela variavel price para termos como exemplo, iremos utilizar o gráfico do histograma para verificar onde os dados estão mais concentrados na variável para poder comparar a população e as amostras

```python
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(12,8))
fig.suptitle('Distribuição dos preços na população e amostras')

sns.histplot(ax=axes[0,0],data=df,x='price',kde=True)
axes[0,0].set_title('População')

sns.histplot(ax=axes[0,1],data=amostra1,x='price',kde=True)
axes[0,1].set_title('Amostra 1')

sns.histplot(ax=axes[1,0],data=amostra2,x='price',kde=True)
axes[1,0].set_title('Amostra 2')

sns.histplot(ax=axes[1,1],data=amostra3,x='price',kde=True)
axes[1,1].set_title('Amostra 3')
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(12,8))
fig.suptitle('Boxplot preços na população e amostras')

sns.boxplot(ax=axes[0,0],data=df,x='price')
axes[0,0].set_title('População')

sns.boxplot(ax=axes[0,1],data=amostra1,x='price')
axes[0,1].set_title('Amostra 1')

sns.boxplot(ax=axes[1,0],data=amostra2,x='price')
axes[1,0].set_title('Amostra 2')

sns.boxplot(ax=axes[1,1],data=amostra3,x='price')
axes[1,1].set_title('Amostra 3')
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(12,8))
fig.suptitle('Distribuição do cut na população e amostras')

sns.histplot(ax=axes[0,0],data=df,x='cut',stat='percent')
axes[0,0].set_title('População')

sns.histplot(ax=axes[0,1],data=amostra1,x='cut',stat='percent')
axes[0,1].set_title('Amostra 1')

sns.histplot(ax=axes[1,0],data=amostra2,x='cut',stat='percent')
axes[1,0].set_title('Amostra 2')

sns.histplot(ax=axes[1,1],data=amostra3,x='cut',stat='percent')
axes[1,1].set_title('Amostra 3')
```
